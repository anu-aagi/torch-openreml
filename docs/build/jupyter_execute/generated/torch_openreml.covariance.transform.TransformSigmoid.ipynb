```python
import torch
from torch_openreml.covariance.transform import TransformSigmoid

t = TransformSigmoid()
x = torch.tensor([-2.0, 0.0, 2.0])
t(x)
```

```python
import torch
from torch_openreml.covariance.transform import TransformSigmoid

t = TransformSigmoid()
x = torch.tensor([0.1, 0.5, 0.9])
t.inverse(x)
```

```python
import torch
from torch_openreml.covariance.transform import TransformSigmoid

t = TransformSigmoid()
x = torch.tensor([0.0, 1.0])
t.grad(x)
```
