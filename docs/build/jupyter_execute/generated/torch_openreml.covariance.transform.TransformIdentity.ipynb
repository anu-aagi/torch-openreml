```python
import torch
from torch_openreml.covariance.transform import TransformIdentity

t = TransformIdentity()
x = torch.tensor([0.0, 1.0, -3.5])
t(x)
```

```python
import torch
from torch_openreml.covariance.transform import TransformIdentity

t = TransformIdentity()
x = torch.tensor([2.0, -1.0])
t.inverse(x)
```
